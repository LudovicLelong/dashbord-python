```python
import pandas as pd
import datetime

# Chargement des données dans un dataframe
array = pd.read_csv("./details_agence_dateformat_age_ref_et_date_fin_vie_version4.csv")


#changer type colonne
array['Annee_RefT'] = pd.to_datetime(array['Annee_RefT'], format='%Y')
array["Date_Fin_Vie"] = pd.to_datetime(array['Date_Fin_Vie'], format='%d/%m/%Y')



# Convertir la colonne "Date_Fin_Vie" en un format de date utilisable par Python
array["Date_Fin_Vie"] = pd.to_datetime(array["Date_Fin_Vie"], errors='coerce')

# Remplacer les valeurs manquantes par -1
array["Date_Fin_Vie"].fillna(-1, inplace=True)

# Créer une variable date_reference correspondant à la date de référence
date_reference = datetime.date(2023, 5, 7)

# Filtrer les lignes pour lesquelles la date de fin de vie est égale à -1
array = array[array["Date_Fin_Vie"] != -1]

# Calcul de l'âge du matériel en années
array["Age_du_materiel"] = (date_reference - array["Date_Fin_Vie"]).dt.days / 365

# Affichage des résultats
print(array[["Materiel", "Date_Fin_Vie", "Age_du_materiel"]])
```
